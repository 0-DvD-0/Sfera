```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_histo_2D(dat_PATH):
    data = np.loadtxt(dat_PATH, skiprows=0, comments='==')
    X = np.zeros((int(data.shape[0]/2), data.shape[1]))       #np.array([[]] * int(data.shape[0]/2))
    Y = np.zeros((int(data.shape[0]/2), data.shape[1]))       #np.array([[]] * int(data.shape[0]/2))

    for i in range(int(data.shape[0]/2)):
        X[i] = (data[i])
        Y[i] = (data[i + 16])

    return X, Y
```

```python
PATH_SAM="../Dati/10-05/Run_ch_amp_XYHistos_5_10_2024.dat"
X, Y = load_histo_2D(PATH_SAM)
```

```python
ch = 15
plt.scatter(X[ch, :], Y[ch, :], marker = ".")
plt.xlim([-4900, 0])
```
