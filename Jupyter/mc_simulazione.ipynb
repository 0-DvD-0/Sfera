```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def monte_carlo(N, xmin, xmax, ymax):
    rng=np.random.default_rng(624374)
    check=True
    array=[]
    while check:
        x=rng.uniform(xmin, xmax)
        y=rng.uniform(0, ymax)
        if y<(11.931*np.exp(-x/247)+8313*np.exp(-x/3.91)):
            array.append(x)
        if len(array)==N:
            check=False
    array+=rng.normal(0,1,N)*2.5
    return np.array(array)
            
```

```python
N=23856
mc=monte_carlo(N, 0, 125, 700 )
```

```python
_=plt.hist(mc, bins=np.arange(0, 125, step=2.5))
```
