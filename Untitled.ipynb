```python
from scipy.signal import convolve as conv
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy.integrate as integral
from scipy.stats import norm
from scipy.fft import fft, ifft
```

```python
def gauss(E, mu, sigma):
    return 1.0 / (np.sqrt(2.0 * np.pi) * sigma) * np.exp(-np.power((E - mu) / sigma, 2.0) / 2)
```

```python
def spectrum (E):
    f1=np.zeros_like(E)
    
    mask = (0.1 < E) & (E < 0.95)  # Create a mask for E values in range (0.1, 0.95)
   
    f1[mask]=1/E[mask]
    return f1 + 0.5*gauss(E, 1.2, 0.01) + 0.1*gauss(E, 0.5, 0.01)
def resolution (E, sigma):
    return gauss(E, 0, np.sqrt(E)*sigma)
```
